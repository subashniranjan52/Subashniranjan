```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(42)

# Generate random dates for a month
dates = pd.date_range(start="2025-01-01", end="2025-01-31", freq='D')

# Simulate random petroleum stock prices
initial_price = 50  # Start with a base price
price_changes = np.random.normal(0, 1.5, len(dates))  # Random daily change
prices = initial_price + np.cumsum(price_changes)

# Create a DataFrame
data = pd.DataFrame({
    'Date': dates,
    'Price': prices
})

# Calculate a moving average (7-day)
data['7-Day MA'] = data['Price'].rolling(window=7).mean()

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(data['Date'], data['Price'], label="Petroleum Stock Price", color='blue')
plt.plot(data['Date'], data['7-Day MA'], label="7-Day Moving Average", color='orange', linestyle='--')
plt.title('Simulated Petroleum Stock Market Data')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()

# Save to CSV (optional)
data.to_csv('petroleum_stock_data.csv', index=False)
```
